```python
from selenium.common.exceptions import StaleElementReferenceException, NoSuchElementException
from selenium import webdriver
from selenium.webdriver.chrome.service import Service
from selenium.webdriver.common.by import By
import time
import pandas as pd
```

```python
service = Service(executable_path="../chromedriver.exe")
options = webdriver.ChromeOptions()
driver = webdriver.Chrome(service=service, options=options)
```

```python
url='https://www.amazon.com/s?k=redmi+earbuds&crid=3TZZRJIPFBP2E&sprefix=redmi+earbuds%2Caps%2C380&ref=nb_sb_noss_1'
driver.get(url)
```

```python
amazon_reviews=[]
count = 0
max_reviews=350
base_url='https://amazon.com'
```

```python
all_products = driver.find_elements(By.XPATH, f"//a[@class='a-link-normal s-underline-text s-underline-link-text s-link-style a-text-normal']")
all_products_links=[]
for product in all_products:
    product_url = product.get_attribute("href")
    # Ensure the URL is correctly formatted
    if not product_url.startswith('https'):
        product_url = base_url + product_url
    all_products_links.append(product_url)
```

```python
# Fetch all products initially

for product_url in all_products_links:
    try: 
        driver.get(product_url)  # Navigate to each product page
        time.sleep(2)

        try:
            # Click on 'See All Reviews' button
            see_all_button = driver.find_element(By.XPATH, f"//a[@data-hook='see-all-reviews-link-foot']")
            see_all_button.click()
            print('See All Reviews clicked')
            time.sleep(2)

            while count < max_reviews:
                all_reviews = driver.find_elements(By.XPATH, f"//span[@data-hook='review-body']")
                for review in all_reviews:
                    amazon_reviews.append(review.text.strip())
                    count += 1
                    print(f"Scraped {count} reviews.")

                    if count >= max_reviews:
                        print(f"Reached the limit of {max_reviews} reviews.")
                        break
                
                if count >= max_reviews:
                    break
                
                try:
                    next_button = driver.find_element(By.XPATH, f"//li[@class='a-last']")
                    next_link = next_button.find_element(By.TAG_NAME, "a")
                    next_url = next_link.get_attribute("href")

                    # Navigate to the next review page
                    driver.get(next_url)
                    print('Next button clicked successfully')
                    time.sleep(2)

                except NoSuchElementException:
                    print('No more pages for this product.')
                    break  # Exit the loop if there is no "Next" button

        except NoSuchElementException as e:
            print(f"Error finding See All Reviews button: {e}")
            amazon_reviews.append("")
        
    except StaleElementReferenceException as e:
        print(f"StaleElementReferenceException encountered: {e}")
        # Re-fetch all product links in case of a stale reference
        all_products = driver.find_elements(By.XPATH, f"//a[@class='a-link-normal s-underline-text s-underline-link-text s-link-style a-text-normal']")
        continue  # Skip to the next product in the list

    if count >= max_reviews:
        print(f"Reached the limit of {max_reviews} reviews.")
        break

driver.quit()
```
